```python
import pandas as pd
```

```python
student_num= pd.read_excel("student_num.xlsx",header=1,skipfooter=4)
# header=1부터 해야 엑셀의 첫번쨰 줄인 컬럼명 다음줄인 2번째 줄부터 데이터로 들어간다.
# usecols='A:G'   A부터 G열까지의 데이터만 가져온다. 
# skipfooter=4  밑에서부터 4개의 행은 제외하고 가져온다.
```

```python
student_num.head(10)
# 컬렴명에 '??.1'형식으로 되어 있는 컬럼은 제거하자! 
```

```python
student_num_2 = student_num[(student_num['교육청']=='계')&(student_num['설립']=='소계')]
student_num_2 
```

```python
# 전국 합계와 각 도시의 총합이 일치하는 지 확인
student_num_2[(student_num_2 ['교육청']=='계')&(student_num_2 ['설립']=='소계')]['계']

# 리스트로 변환!
list_student = list(student_num_2[(student_num_2['교육청']=='계')&(student_num_2['설립']=='소계')]['계'])

list_student[0] == sum(list_student[1:])
```

```python
# 열 이름이 ''??.1' ,  '??.2' , '??.3'   형태인 것은 삭제! 

student_num_2.columns
```

```python
# 컬럼이름을 받아서 이름안에 1, 2 혹은 3이 있으면 그 열은 삭제

for i in student_num_2.columns:
    if str(1) in i or str(2) in i or str(3) in i:
        student_num_2.drop(i, axis=1, inplace=True)

student_num_2.reset_index(drop=True,inplace=True) 
student_num_2
```

```python
student_num_2.shape     # 원래 열길이가 29개 였는데 제거 후 15개가 되었다. 
```

```python
# '-' 는 0으로 처리
student_num_2 = student_num_2.replace('-',0)
```

```python
student_num_2
```

```python
# '교육청','설립' 컬럼 삭제
student_num_2.drop(['교육청','설립'], axis=1, inplace=True)
student_num_2
```

```python
# 학교형태별로 전국 = 나머지합 이 되는지 확인
for i in range(len(student_num_2.columns)-1):
    print(list(student_num_2.iloc[:,i+1])[0] ==  sum(list(student_num_2.iloc[:,i+1])[1:]))
    
    
    
# for문이 돌아가는 범위에 -1을 한 이유는 '시도'컬럼은 합계를 계산하지 않기 때문에 제외되었기 때문에 1을 뺸 것이다.
# student_num_2.iloc를 이용해 열로 접근. 계산을 시작하는 열은  0번재열인 '시도'가  아닌 1번째 열인 '계' 이다.
# 근데 열을 i로 받게되면 0번째 열부터 계산이 되기 때문에 i+1을 하였다. 
# 이렇게 열로 접근한 것을 리스트로 받아 리스트내의 0번째 방, 즉 전국
# 그리고 1번째 방부터 끝까지의 합을 구해서 둘이 같은지 확인. 모두 true가 나오면 성공!!
```

```python
# 지역별로 '계'= 학교별 합이 되는지 확인

for i in range(len(student_num_2)):
    print(list(student_num_2.loc[i,])[1] == sum(list(student_num_2.loc[i,])[2:]))
    
    
# 행길이 만큼 for문을 돌려서 지역별로 합을 확인
# loc(행)으로 접근. 0번째 행부터 계산. 그 것을 리스트로 만든다. 그러면 리스트의 0번째에는 각 '시도'이름이 오기 때문에
# 리스트의 1번 째 방('계') == 나머지들의 합과 같은지 확인. true이면 성공!!
```

```python
# 정리한 student_num_2를 csv로 저장한다. 
student_num_2 

student_num_2.to_csv('student_num_2.csv',index=False,encoding='utf-8-sig')
```
