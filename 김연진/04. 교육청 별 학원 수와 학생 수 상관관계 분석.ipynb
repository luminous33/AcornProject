```python
import pandas as pd
import numpy as np

import matplotlib.font_manager as fm            # 폰트 지정방법
font_name=fm.FontProperties(fname='C:\\Windows\\Fonts\\malgun.ttf').get_name()     
plt.rc('font',family=font_name)
```

```python
hakwon = pd.read_csv('hakwon_stu.csv')
```

```python
hakwon
```

```python
#'교육청'컬럼 없애고 '시도'를 앞으로
hakwon.drop('교육청',axis=1,inplace=True)
hakwon = hakwon[['시도','학원 수','학원 당 학생수의 평균','일시수용능력인원합계','학원 설립연도 평균','초등학교','중학교','일반고','특목고']]
hakwon
```

#### 시각화

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(7, 7))         # 플롯 사이즈 지정
plt.scatter(hakwon['학원 수'],hakwon['초등학교']) 
plt.xlabel("학원 수")                  # x축 레이블 지정
plt.ylabel("초등학생 수")                   # y축 레이블 지정
plt.grid()                           # 플롯에 격자 보이기
for i in range(17):
    plt.text(hakwon['학원 수'][i],hakwon['초등학교'][i],hakwon['시도'][i],fontsize=11)


plt.show()                           # 플롯 보이기 
```

##### 학원수와 학생수 for문

```python
school_list =['초등학교','중학교','일반고','특목고']

for i in school_list:
    plt.figure(figsize=(6, 6))         # 플롯 사이즈 지정
    plt.scatter(hakwon['학원 수'],hakwon[i]) 
    plt.title('{} 학생 수와 학원 수'.format(i))
    plt.xlabel("학원 수")                  # x축 레이블 지정
    plt.ylabel("{} 학생 수".format(i))                   # y축 레이블 지정
    plt.grid()                           # 플롯에 격자 보이기
    
    for j in range(17):
        plt.text(hakwon['학원 수'][j],hakwon[i][j],hakwon['시도'][j],fontsize=11)


plt.show()    
```

#### 도시별 학생수

```python
plt.barh(hakwon['시도'], hakwon['초등학교'])
plt.title('초등학생 수')
plt.xlabel("도시")                  # x축 레이블 지정
plt.ylabel("학생 수")                   # y축 레이블 지정
```

```python
school_list =['초등학교','중학교','일반고','특목고']

for i in school_list:
    plt.barh(hakwon['시도'], hakwon[i])
    plt.title('{}학생 수'.format(i))
    plt.xlabel("도시")                  # x축 레이블 지정
    plt.ylabel("학생 수")                   # y축 레이블 지정
    
    


    plt.show()  ###### for문 이용
```
