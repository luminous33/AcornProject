```python
import pandas as pd
import numpy as np
```

```python
hakwon_edu = pd.read_csv('hakwon_edu.csv')
student_num = pd.read_csv('student_num_2.csv')
```

```python
hakwon_edu
```

```python
student_num
```

```python
# student_num에 있는 전국은 일단 제외
student_num_2 = student_num.loc[1:,]
```

```python
student_num_2
```

```python
# student_num_2도 미리 인덱스 순서 지정하기
student_num_2.reset_index(inplace=True)
student_num_2.set_index('시도',inplace=True)
```

```python
student_num_2.sort_index(ascending=True,inplace=True)
student_num_2
```

```python
student_num_2
```

```python
# 열도 학원을 많이 다닐 것 같은 학교만 남기자(초등,중등,일반고,특목고)

student_num_3 = student_num_2.loc[:,['시도','초등학교','중학교','일반고','특목고']]
```

```python
# 시도를 인덱스로
student_num_3.set_index('시도',inplace=True)
student_num_3
```

```python
# student_num 인덱스를 학원과 맞추자. 
# 학원인덱스가 가나다 순이어서 student_num 인덱스도 sort_index를 이용해 가나다 순으로 정렬
student_num_3.sort_index(ascending=True,inplace=True)
```

```python
# 학원데이터에 그래도 student_num을 merge시킨다. 
#  pd.merge(hakwon_edu,student_num_3)   -> 그냥 merge를 시키니까 공통 컬럼에 없다고 에러가 난다. 
# '시도'라는 공통컬럼만들기
# 그래서 student_num_3을 인덱스를 리셋시킨다

student_num_3.reset_index(inplace=True,drop=False)
student_num_3
```

```python
# hakwon도 '시도'라는 컬럼 만들기
hakwon_edu['시도'] = student_num_3['시도']

hakwon_edu
```

```python
# merge 하기!

hakwon_stu = pd.merge(hakwon_edu,student_num_3)
hakwon_stu
```

```python
# 저장!!

hakwon_stu.to_csv('hakwon_stu.csv',encoding='utf-8-sig',index=False)
```
